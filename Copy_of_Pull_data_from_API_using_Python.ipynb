```python
import requests
import pandas
import sqlalchemy
```

```python
url = 'http://api.coincap.io/v2/assets'
header ={"Content-Type":"application/jason","Accept-Encoding":"deflate"}

response = requests.get(url, headers=header)

print(response)
```

```python
responseData = response.json()

#print(responseData)
```

```python
df=pandas.json_normalize(responseData, "data")
print(df)
```

```python
engine=sqlalchemy.create_engine("mssql+pyodb://servername/databasename?driver=SQL+Server+Native+Client11.0")
df.to_sql(name="Tablename",con=engine,index="False|Replace|Apend")
```
